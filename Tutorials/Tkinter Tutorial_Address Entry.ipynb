```python
import tkinter as tk

# Create a window with title "Address Entry Form"
window = tk.Tk()
window.title("Address Entry Form")

# Create new frame "frame_form" to contain the Label and Entry widgets
frame_form = tk.Frame(relief = tk.SUNKEN, borderwidth = 3)
frame_form.pack()

# Create Label and Entry form
lbl_first_name = tk.Label(master = frame_form, text = "First Name:")
ent_first_name = tk.Entry(master = frame_form, width = 50)
lbl_last_name = tk.Label(master = frame_form, text = "Last Name:")
ent_last_name = tk.Entry(master = frame_form, width = 50)
lbl_add_line1 = tk.Label(master = frame_form, text = "Address Line 1:")
ent_add_line1 = tk.Entry(master = frame_form, width = 50)
lbl_add_line2 = tk.Label(master = frame_form, text = "Address Line 2:")
ent_add_line2 = tk.Entry(master = frame_form, width = 50)
lbl_city = tk.Label(master = frame_form, text = "City:")
ent_city = tk.Entry(master = frame_form, width = 50)
lbl_state = tk.Label(master = frame_form, text = "State/Province:")
ent_state = tk.Entry(master = frame_form, width = 50)
lbl_postal = tk.Label(master = frame_form, text = "Postal Code:")
ent_postal = tk.Entry(master = frame_form, width = 50)
lbl_country = tk.Label(master = frame_form, text = "Country:")
ent_country = tk.Entry(master = frame_form, width = 50)

# Placing of Label and Entry widgets
lbl_first_name.grid(row = 0, column = 0, sticky = "e")
ent_first_name.grid(row = 0, column = 1)
lbl_last_name.grid(row = 1, column = 0, sticky = "e")
ent_last_name.grid(row = 1, column = 1)
lbl_add_line1.grid(row = 2, column = 0, sticky = "e")
ent_add_line1.grid(row = 2, column = 1)
lbl_add_line2.grid(row = 3, column = 0, sticky = "e")
ent_add_line2.grid(row = 3, column = 1)
lbl_city.grid(row = 4, column = 0, sticky = "e")
ent_city.grid(row = 4, column = 1)
lbl_state.grid(row = 5, column = 0, sticky = "e")
ent_state.grid(row = 5, column = 1)
lbl_postal.grid(row = 6, column = 0, sticky = "e")
ent_postal.grid(row = 6, column = 1)
lbl_country.grid(row = 7, column = 0, sticky = "e")
ent_country.grid(row = 7, column = 1)

# Create a new frame "frame_buttons" to contain the submit and clear buttons
frame_buttons = tk.Frame()
frame_buttons.pack(fill = tk.X, ipadx = 5, ipady = 5)

# Create "Submit" button and pack it to right side of the frame
button_submit = tk.Button(master = frame_buttons, text = "Submit")
button_submit.pack(side = tk.RIGHT, padx = 10, ipadx = 10)

# Create "Clear" button and pack it to the right side of the frame
button_clear = tk.Button(master = frame_buttons, text = "Clear")
button_clear.pack(side = tk.RIGHT, ipadx = 10)

window.mainloop()
```

```python
# Smarter way

import tkinter as tk

# Create a new window with the title "Address Entry Form"
window = tk.Tk()
window.title("Address Entry Form")

# Create a new frame `frm_form` to contain the Label
# and Entry widgets for entering address information.
frm_form = tk.Frame(relief=tk.SUNKEN, borderwidth=3)
# Pack the frame into the window
frm_form.pack()

# List of field labels
labels = [
    "First Name:",
    "Last Name:",
    "Address Line 1:",
    "Address Line 2:",
    "City:",
    "State/Province:",
    "Postal Code:",
    "Country:",
]

# Loop over the list of field labels
for idx, text in enumerate(labels):
    # Create a Label widget with the text from the labels list
    label = tk.Label(master=frm_form, text=text)
    # Create an Entry widget
    entry = tk.Entry(master=frm_form, width=50)
    # Use the grid geometry manager to place the Label and
    # Entry widgets in the row whose index is idx
    label.grid(row=idx, column=0, sticky="e")
    entry.grid(row=idx, column=1)

# Create a new frame `frm_buttons` to contain the
# Submit and Clear buttons. This frame fills the
# whole window in the horizontal direction and has
# 5 pixels of horizontal and vertical padding.
frm_buttons = tk.Frame()
frm_buttons.pack(fill=tk.X, ipadx=5, ipady=5)

# Create the "Submit" button and pack it to the
# right side of `frm_buttons`
btn_submit = tk.Button(master=frm_buttons, text="Submit")
btn_submit.pack(side=tk.RIGHT, padx=10, ipadx=10)

# Create the "Clear" button and pack it to the
# right side of `frm_buttons`
btn_clear = tk.Button(master=frm_buttons, text="Clear")
btn_clear.pack(side=tk.RIGHT, ipadx=10)

# Start the application
window.mainloop()
```
